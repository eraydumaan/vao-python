```python
def fibonacci_sequence(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    else:
        list_fib = [0, 1]
        while len(list_fib) < n:
            next_fib = list_fib[-1] + list_fib[-2]
            list_fib.append(next_fib)
        return list_fib


fib_series = fibonacci_sequence(20)
print(f"Fibonacci Dizisi (en fazla 20 terim):\n{fib_series}")
```
